# Hipotesis:

Hipótesis nula (H₀):
No existe una relación estadísticamente significativa entre las variables masa corporal, tensión arterial, peso y medida de cintura con el riesgo de hipertensión.

Hipótesis alternativa (H₁):
Existe al menos una relación estadísticamente significativa entre alguna de las variables masa corporal, tensión arterial, peso o medida de cintura y el riesgo de hipertensión.


```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import matplotlib.pyplot as plt

X = df[[ 'peso', 'medida_cintura', 'tension_arterial', 'masa_corporal']]
Y = df['riesgo_hipertension']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42 )

x_train_constante = sm.add_constant(X_train)
x_test_constante = sm.add_constant(X_test)

modelo = sm.OLS(Y_train, x_train_constante).fit()

Yc = modelo.predict(x_test_constante)

plt.scatter(Y_test, Yc)

r2 = r2_score(Y_test, Yc)
print(f'Coeficiente de determinación: {r2: 0.4}')

from statsmodels.formula.api import ols

moedo_2 = ols(formula = 'riesgo_hipertension ~ peso + medida_cintura + tension_arterial + masa_corporal', data = df).fit()
tabla_anova = sm.stats.anova_lm(moedo_2, typ = 2)  
tabla_anova
```

```python
b0, b1, b2, b3, b4,  = modelo.params

Fun = lambda X1, X2, X3, X4: b0 + b1 * X1 + b2 * X2 + b3 * X3 + b4 * X4 

#Eres el profesional de salud responsable de valorar a un paciente que presenta sospechas de hipertensión arterial. Durante el examen médico inicial, se recopilan datos relevantes relacionados con su estado físico y fisiológico.
#Con base en esta información recolectada, se procede a analizar:
#¿Existe una alta probabilidad de que el paciente presente hipertensión?
Fun(118, 135, 110, 35.22)
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

stats.probplot(modelo.resid, dist="norm", plot=plt)
plt.title( 'Redisuales')
plt.show()

plt.hist(modelo.resid, bins=30)
plt.title("Histograma de los residuos")
plt.xlabel("Residuo")
plt.ylabel("Frecuencia")
plt.show()
```

```python
from scipy.stats import shapiro

stat, p = shapiro(modelo.resid)
print('Estadístico Shapiro-Wilk=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
    print("Los residuos parecen normales.")
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan

_, pval, __, ___ = het_breuschpagan(modelo.resid, modelo.model.exog)
print("p-valor de Breusch-Pagan:", pval)
```

# Interpretación general de los resultados
A partir del análisis de varianza (ANOVA), se examinó el efecto de cuatro variables predictoras —peso, medida de cintura, tensión arterial y masa corporal— sobre una variable dependiente asociada al riesgo de hipertensión. El modelo obtuvo un coeficiente de determinación (R²) de 0.3059, lo cual indica que aproximadamente el 30.59% de la variabilidad de la variable dependiente es explicada por las variables independientes incluidas en el modelo.

Los resultados reflejan que tres de las cuatro variables analizadas presentan una asociación estadísticamente significativa con el riesgo de hipertensión, evidenciado por valores-p menores al nivel de significancia convencional (α = 0.05), a saber:

Medida de cintura (p = 1.27e-05)

Tensión arterial (p = 2.09e-60)

Masa corporal (p = 4.90e-24)

En contraste, la variable peso (p = 0.3146) no mostró un efecto estadísticamente significativo dentro del modelo.

Dado que al menos una de las variables independientes presenta una relación significativa con la variable dependiente, se procede a rechazar la hipótesis nula (H₀). Por tanto, se concluye que existe evidencia estadística suficiente para afirmar que al menos una de las variables incluidas en el modelo influye significativamente en el riesgo de hipertensión en la población estudiada.

# Interpretación del modelo aplicado al caso clínico
Al ingresar los valores clínicos de un paciente al modelo ajustado, se observó una alta probabilidad de hipertensión, lo cual confirma la sospecha clínica inicial. Esta predicción sugiere que el paciente podría encontrarse en una situación de riesgo elevado o incluso ya estar cursando un cuadro hipertensivo, lo que justifica una evaluación médica más profunda.

# Evaluación de supuestos del modelo
 Normalidad de los residuos
Se aplicó la prueba de Shapiro-Wilk para evaluar si los residuos del modelo siguen una distribución normal. Dado que el valor-p fue inferior a 0.05 (p = 0.000), se rechaza la hipótesis nula, concluyéndose que los residuos no siguen una distribución normal. Esto viola el supuesto de normalidad, fundamental en modelos de regresión lineal clásica para asegurar la validez de los intervalos de confianza y las pruebas de significancia.

 Homocedasticidad
La homocedasticidad —es decir, la constancia de la varianza de los errores— fue evaluada mediante una prueba específica. Dado que el valor-p fue cercano a cero, se rechaza la hipótesis nula, lo que indica la presencia de heterocedasticidad. Esta condición supone que la varianza de los residuos varía a lo largo del rango de valores predichos, lo cual compromete la eficiencia de los estimadores del modelo.

# Conclusión general
El análisis estadístico permitió identificar factores clínicos significativamente asociados al riesgo de hipertensión. Se encontró que la medida de cintura, la tensión arterial y la masa corporal son variables predictoras relevantes, mientras que el peso corporal no mostró un efecto estadísticamente significativo.

Aunque el modelo explica un porcentaje moderado de la variabilidad de la variable dependiente (R² = 0.3059), lo cual es aceptable en contextos biomédicos donde múltiples factores interactúan, se identificaron violaciones importantes a los supuestos del modelo de regresión lineal clásica, específicamente:

No normalidad de los residuos (Shapiro-Wilk, p < 0.05)

Heterocedasticidad en los errores (p ≈ 0)

Estas condiciones pueden comprometer la precisión de las inferencias del modelo, por lo que se recomienda considerar ajustes metodológicos, como el uso de modelos de regresión robusta, transformaciones de variables o métodos no paramétricos.

Finalmente, la aplicación del modelo a un caso clínico real arrojó una alta probabilidad de hipertensión, lo cual respalda su utilidad como herramienta predictiva complementaria, siempre y cuando se usen con criterio y atendiendo a sus limitaciones estadísticas.
